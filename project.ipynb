Setting the model

```python

depth_values = range(1, 21)  # Adjust range based on your dataset
mean_scores = []

# Loop through each max_depth and evaluate using cross-validation
for depth in max_depth_values:
    model = DecisionTreeClassifier(max_depth=depth, random_state=42)
    scores = cross_val_score(model, X_train, y_train, cv=5)  # 5-fold cross-validation
    mean_scores.append(scores.mean())

# Find the best max_depth with the highest mean cross-validation score
best_depth = max_depth_values[mean_scores.index(max(mean_scores))]
print(f"Best max_depth: {best_depth} with accuracy: {max(mean_scores):.4f}")
```

Plot the max_depth with accuracy


```python
plt.plot(max_depth_values, mean_scores, marker='o')
plt.xlabel('Max Depth')
plt.ylabel('Cross-Validated Accuracy')
plt.title('Decision Tree Accuracy for Different Max Depths')
plt.show()
```

```python
from sklearn.datasets import make_classification
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt

# Generate synthetic dataset
X, y = make_classification(
    n_samples=1000, n_features=10, n_informative=5, random_state=42
)

# Define feature names
#feature_names = [f'Feature_{i+1}' for i in range(X.shape[1])]
feature_names = ['magnitude','cdi','alert','sig','nst','magType','depth','latitude','longitude','Year']
# Fit a Decision Tree model
model = DecisionTreeClassifier(random_state=42)
model.fit(X, y)

# Get feature importance
importance = model.feature_importances_

# Print feature importance
print("Feature Importance:")
for name, score in zip(feature_names, importance):
    print(f"{name}: {score:.5f}")

# Plot feature importance
plt.figure(figsize=(10, 6))
plt.bar(feature_names, importance, color="skyblue")
plt.xticks(rotation=45, fontsize=10)
plt.title("Feature Importance from Decision Tree", fontsize=14)
plt.xlabel("Feature", fontsize=12)
plt.ylabel("Importance Score", fontsize=12)
plt.tight_layout()  # Adjust layout to prevent label overlap
plt.show()
```
